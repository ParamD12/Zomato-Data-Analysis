```python
# Convert Data Type of Order_Date 
orders['Order_Date'] = pd.to_datetime(orders['Order_Date'])
```

```python
orders.info()
```

```python
orders['Month'] = orders['Order_Date'].dt.strftime('%B') # Date-Time Module Format Code - w3Schools

orders
```

```python
orders["Hour"] = orders["Order_Date"].dt.strftime("%H")
orders["Hour"] = orders["Hour"].astype("int")

orders
```

```python
orders['Day_of_Week'] = orders['Order_Date'].dt.day_name()

orders
```

## Exploratory Data Analysis [EDA] & Data Visualization

```python
# Total Revenue Generated 

total_revenue = float(round(orders['Total_Amount'].sum(), 2))

print("Total Revenue -", total_revenue)
```

```python
# Total Number of Orders

total_orders = int(orders['Order_ID'].count())

print("Total Orders -", total_orders)
```

```python
# Average Value Per Order [AOV]

AOV = total_revenue / total_orders

print("Average Value Per Order [AOV] -", AOV)
```

```python
# Average Delivery Time

Average_Delivery_Time = float(round(orders['Actual_Delivery_Time'].sum()/orders['Order_ID'].count(), 2))

print("Average Delivery Time -", Average_Delivery_Time, "Minutes")
```

```python
# Cancellation Rate
 
cancelled_orders = orders[orders['Order_Status'] == 'Cancelled'].shape[0]
cancellation_rate = round((cancelled_orders / total_orders) * 100, 2)

print("Cancellation Rate -", cancellation_rate, "%")

# Create a count plot for Order Status
plt.figure(figsize=(8, 6))

ax = sns.countplot(data=orders, x='Order_Status', order=['Completed', 'Cancelled'], color='#E32636')
ax.bar_label(ax.containers[0])

plt.title('Count of Completed and Cancelled Orders')

plt.xlabel('Order Status')
plt.ylabel('Count')

plt.show()
```

```python
# Number of Orders by Hour

orders_by_hour = orders.groupby('Hour').size()

x_positions = np.arange(len(orders_by_hour)) * 1.0  # Adjust spacing by changing 1.3

plt.figure(figsize=(15, 6))  

# Bar plot for the number of orders per hour
ax = sns.barplot(x = x_positions, y = orders_by_hour.values, color='#E32636', width=0.8)

# Line plot to show the trend
sns.lineplot(x = x_positions, y=orders_by_hour.values, color='#740938', marker='o')

# Add labels on top of each bar
ax.bar_label(ax.containers[0], padding=4)

# Set titles and labels
plt.title('Number of Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')

plt.xticks(ticks=np.arange(len(orders_by_hour)), labels=orders_by_hour.index)

plt.show()
```

```python
# Numbers of Orders by Days of the Week

plt.figure(figsize=(10, 6))

orders_by_day = orders['Day_of_Week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

ax = sns.barplot(x=orders_by_day.index, y=orders_by_day.values, color='#E32636')

ax.bar_label(ax.containers[0], padding=2, fmt='%d')

# Set titles and labels
plt.title('Number of Orders by Day of Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')

plt.show()
```

```python
# Numbers of Orders by Month

plt.figure(figsize=(10,8))

orders_by_month = orders['Month'].value_counts().reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October'])

ax = sns.barplot(x=orders_by_month.index, y=orders_by_month.values, color='#E32636')

ax.bar_label(ax.containers[0], padding = 2)

plt.xticks(rotation = 45) # Text Label on X-Axis

# Set titles and labels
plt.title('Number of Orders by Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')

plt.show()
```

```python
# Revenue Generated Over Months

revenue_by_month = orders.groupby('Month')['Total_Amount'].sum().reindex(
    ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October']
    )

x_positions = np.arange(len(revenue_by_month)) * 1.0

plt.figure(figsize=(12, 8))

ax = sns.barplot(x=x_positions, y=revenue_by_month.values, color='#E32636')

sns.lineplot(x = x_positions, y=revenue_by_month.values, color='#740938', marker='o')

ax.bar_label(ax.containers[0], padding = 5)

# Set titles and labels
plt.title('Revenue Generated By Month')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.xticks(ticks=np.arange(len(revenue_by_month)), labels=revenue_by_month.index)

plt.show()
```

```python
# Top 10 Dishes as per Revenue Contribution

top_selling_dishes_revenue = orders.groupby('Dish_Name')['Total_Amount'].sum().head(10)

plt.figure(figsize=(8, 8))

colors = sns.color_palette("Reds", len(top_selling_dishes_revenue))  # Use red hues for the pie chart

# Pie chart for top 10 dishes by revenue contribution
plt.pie(top_selling_dishes_revenue, labels=top_selling_dishes_revenue.index, autopct='%1.1f%%', startangle=140, colors=colors)

plt.title('Top 10 Dishes by Revenue Contribution')
plt.show()
```

```python
# Top 10 Dishes as per Number of Orders

top_selling_dishes = orders.groupby('Dish_Name')['Order_ID'].count().head(10)

num_orders_top_10 = len(top_selling_dishes)

plt.figure(figsize=(8, 8))

ax = sns.barplot(x=top_selling_dishes.values, y=top_selling_dishes.index, color='#E32636')

ax.bar_label(ax.containers[0], padding = 2)

plt.xlabel('Number of Orders')
plt.ylabel('Dish Name')

plt.title('Top 10 Dishes by Number of Orders')

plt.show()
```

```python
orders_with_location = pd.merge(orders, customers[['Customer_ID', 'Customer_Location']], on='Customer_ID', how='left')

orders_with_location
```

```python
# Top 5 Dishes as per Location 
 
# Group by location and dish name to find the count of orders for each dish
top_dishes = orders_with_location.groupby(['Customer_Location', 'Dish_Name']).size().reset_index(name='Order_Count')

# Find the top 5 dishes for each location
top_5_dishes_per_location = top_dishes.sort_values(['Customer_Location', 'Order_Count'], ascending=[True, False]).groupby('Customer_Location').head(5)

# Prepare the data for plotting
locations = top_5_dishes_per_location['Customer_Location'].unique()

for location in locations:
    location_data = top_5_dishes_per_location[top_5_dishes_per_location['Customer_Location'] == location]
    
    plt.figure(figsize=(5, 5))
    bars = plt.bar(location_data['Dish_Name'], location_data['Order_Count'], color='#E32636')
    
    plt.bar_label(bars, padding=2)
    
    plt.xlabel('Dish Name')
    plt.ylabel('Number of Orders')

    plt.xticks(rotation = 45)
    plt.title(f'Top 5 Dishes in {location}')
    
    # Display the plot
    plt.show()
```

```python
merged_data = orders.merge(restaurants, on='Restaurant_ID').merge(customers, on='Customer_ID')

merged_data
```

```python
merged_data.columns
```

```python
# Top 5 Cuisines as per Location

top_cuisines = merged_data.groupby(['Customer_Location', 'Cuisine_Types']).size().reset_index(name='Order_Count')

top_5_cuisines_per_city = top_cuisines.sort_values(['Customer_Location', 'Order_Count'], ascending=[True, False]).groupby('Customer_Location').head(5)

cities = top_5_cuisines_per_city['Customer_Location'].unique()

for city in cities:
    city_data = top_5_cuisines_per_city[top_5_cuisines_per_city['Customer_Location'] == city]
    
    plt.figure(figsize=(5, 5))
    bars = plt.bar(city_data['Cuisine_Types'], city_data['Order_Count'], color='#E32636')
    
    plt.bar_label(bars, padding=1)
    
    plt.xlabel('Cuisine Type')
    plt.ylabel('Number of Orders')
    plt.title(f'Top 5 Cuisines in {city}')

    plt.xticks(rotation = 45)
    
    plt.show()
```

```python
# Top 10 Cuisines as per Revenue

revenue_per_cuisine = merged_data.groupby('Cuisine_Types')['Total_Amount'].sum().reset_index()
revenue_per_cuisine = revenue_per_cuisine.rename(columns={'Total_Amount': 'Total_Revenue'})

top_10_cuisines_revenue = revenue_per_cuisine.sort_values(by='Total_Revenue', ascending=False).head(10)

colors = sns.color_palette("Reds", len(top_10_cuisines_revenue))

plt.figure(figsize=(10, 7))
plt.pie(top_10_cuisines_revenue['Total_Revenue'], labels=top_10_cuisines_revenue['Cuisine_Types'], autopct='%1.1f%%', startangle=140, colors = colors)

plt.title('Top 10 Cuisines by Revenue Contribution')
plt.show()
```

```python
# Top 10 Cuisines by Number of Orders

top_10_cuisine = merged_data.groupby('Cuisine_Types')['Order_ID'].count().reset_index(name = 'Order_Count')

top_10_cuisine = top_10_cuisine.sort_values(by='Order_Count', ascending=False).head(10)

plt.figure(figsize=(8, 8))

ax = sns.barplot(y=top_10_cuisine['Cuisine_Types'], x=top_10_cuisine['Order_Count'], color='#E32636')

ax.bar_label(ax.containers[0], padding = 2)

plt.xlabel('Number of Orders')
plt.ylabel('Cuisine Name')

plt.title('Top 10 Cuisines by Number of Orders')

plt.show()
```

```python
# Locations as per Number of Orders

order_count_per_location = merged_data.groupby('Customer_Location').size().reset_index(name='Order_Count')

top_10_locations_orders = order_count_per_location.sort_values(by='Order_Count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_locations_orders['Customer_Location'], top_10_locations_orders['Order_Count'], color='#E32636')

plt.bar_label(bars, padding=2)

plt.xlabel('Customer Location')
plt.ylabel('Number of Orders')
plt.title('Top 10 Locations by Number of Orders')

plt.show()
```

```python
# Locations as per Revenue

revenue_per_location = merged_data.groupby('Customer_Location')['Total_Amount'].sum().reset_index()

top_10_locations_revenue = revenue_per_location.sort_values(by='Total_Amount', ascending=False).head(10)

plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_locations_revenue['Customer_Location'], top_10_locations_revenue['Total_Amount'], color='#E32636')

plt.bar_label(bars, padding=2)

plt.xlabel('Customer Location')
plt.ylabel('Revenue')
plt.title('Top 10 Locations by Revenue Contribution')

plt.show()
```

```python
# Number of Orders by Age Group

num_orders_age_group = merged_data.groupby('Customer_Age_Group').size().reset_index(name='Order_Count')

num_orders_age_group = num_orders_age_group.sort_values(by='Customer_Age_Group', ascending=True).head(10)

plt.figure(figsize=(10, 8))
bars = plt.bar(num_orders_age_group['Customer_Age_Group'],num_orders_age_group['Order_Count'], color='#E32636')

plt.bar_label(bars, padding=2)

plt.xlabel('Customer Age Group')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Age Group')

plt.show()
```

```python
# Revenue Contribution per Age Group

revenue_per_age_group = merged_data.groupby('Customer_Age_Group')['Total_Amount'].sum().reset_index()
revenue_per_age_group = revenue_per_age_group.rename(columns={'Total_Amount': 'Total_Revenue'})

colors = ['#c1121f', '#9d0208', '#ef233c', '#f56960']

plt.figure(figsize=(10, 7))
plt.pie(revenue_per_age_group['Total_Revenue'], labels=revenue_per_age_group['Customer_Age_Group'], autopct='%1.1f%%', startangle=140, colors = colors, wedgeprops = {'edgecolor': 'black'})

plt.legend(title = 'Age Group')

plt.title('Revenue Contribution per Age Group')
plt.show()
```

```python
# Number of Orders by Age Group & Location

plt.figure(figsize = (12,5)) 

ax = sns.countplot(x = "Customer_Location", data = merged_data, hue = "Customer_Age_Group", palette= 'Reds', edgecolor = "black")

plt.legend(bbox_to_anchor = (1,1))

for bars in ax.containers:
    ax.bar_label(bars, padding = 2)

plt.xlabel("Customer Location")
plt.ylabel("Number of Orders")

plt.title("Number of Orders by Age Group & Location")

plt.show()
```

```python
# Top 5 Restaurants for each Location based on Number of Orders

top_restaurants_num_orders = merged_data.groupby(['Location', 'Name']).size().reset_index(name='Order_Count')

top_5_restaurants_per_location = top_restaurants_num_orders.sort_values(['Location', 'Order_Count'], ascending=[True, False]).groupby('Location').head(5)

locations = top_5_restaurants_per_location['Location'].unique()

for location in locations:
    location_data = top_5_restaurants_per_location[top_5_restaurants_per_location['Location'] == location]
    
    plt.figure(figsize=(4, 4))
    bars = plt.bar(location_data['Name'], location_data['Order_Count'], color='#E32636')
    
    plt.bar_label(bars, padding=0)
    
    plt.xlabel('Restaurant Name')
    plt.ylabel('Number of Orders')
    plt.title(f'Top 5 Restaurants in {location}')

    plt.xticks(rotation = 90)
    
    plt.show()
```

```python
# Top 5 Restaurants for each Location based on Revenue Contribution

restaurant_revenue = merged_data.groupby(['Location', 'Name'])['Total_Amount'].sum().reset_index(name='Revenue')

top_5_restaurants_location_revenue = restaurant_revenue.sort_values(['Location', 'Revenue'], ascending=[True, False]).groupby('Location').head(5)

locations = top_5_restaurants_location_revenue['Location'].unique()

for location in locations:
    location_data = top_5_restaurants_location_revenue[top_5_restaurants_location_revenue['Location'] == location]
    
    plt.figure(figsize=(5, 5))
    plt.plot(location_data['Name'], location_data['Revenue'], marker='o', color='#E32636', linestyle='-', linewidth=2)
    
    for x, y in zip(location_data['Name'], location_data['Revenue']):
        plt.text(x, y, f'{y:.2f}', ha='left', va='bottom', fontsize=10)
    
    plt.grid(True, linestyle='--', linewidth=0.5)

    plt.xlabel('Restaurant Name')
    plt.ylabel('Revenue Contribution')

    plt.title(f'Top 5 Restaurants by Revenue in {location}')

    plt.xticks(rotation = 90)

    plt.grid(True)

    plt.show()
```

```python
# Top 5 Restaurants for each Location based on Average Rating

avg_rating_per_restaurant = restaurants.groupby(['Location', 'Name'])['Ratings'].mean().reset_index(name='Average_Rating')

top_5_rating_restaurants_per_location = avg_rating_per_restaurant.sort_values(['Location', 'Average_Rating'], ascending=[True, False]).groupby('Location').head(5)

locations = top_5_rating_restaurants_per_location['Location'].unique()

for location in locations:
    location_data = top_5_rating_restaurants_per_location[top_5_rating_restaurants_per_location['Location'] == location]
    
    plt.figure(figsize=(5, 5))
    bars = sns.barplot(x='Name', y='Average_Rating', data=location_data, color='#E32636')
    
    bars.bar_label(bars.containers[0], fmt='%.1f', padding=0)
    
    plt.xlabel('Restaurant Name')
    plt.ylabel('Average Customer Rating')

    plt.title(f'Top 5 Restaurants by Average Rating in {location}')

    plt.xticks(rotation = 45)

    plt.tight_layout()
    plt.show()
```

```python
# Top 5 Restaurants for each Location based on Number of Reviews

reviews_per_restaurant = merged_data.groupby(['Location', 'Name'])['Reviews_Count'].sum().reset_index(name='Total_Reviews')

top_5_reviewed_restaurants_per_location = reviews_per_restaurant.sort_values(['Location', 'Total_Reviews'], ascending=[True, False]).groupby('Location').head(5)

locations = top_5_reviewed_restaurants_per_location['Location'].unique()

# Plot top 5 restaurants by number of reviews for each location
for location in locations:
    location_data = top_5_reviewed_restaurants_per_location[top_5_reviewed_restaurants_per_location['Location'] == location]
    
    plt.figure(figsize=(10, 5))
    bars = sns.barplot(x='Name', y='Total_Reviews', data=location_data, color='#E32636')
    
    bars.bar_label(bars.containers[0], fmt='%d', padding=0)
    
    plt.xlabel('Restaurant Name')
    plt.ylabel('Number of Reviews')
    plt.title(f'Top 5 Restaurants by Number of Reviews in {location}')
    
    plt.tight_layout()
    plt.show()
```
